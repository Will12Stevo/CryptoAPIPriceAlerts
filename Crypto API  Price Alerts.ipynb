```python
#Setting up automated emails

import os
import requests
from time import sleep
import smtplib
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText
```

```python
# Function to fetch data from the API
def api_runner():
    global df
    
    api_key = '6fae45d7-aea7-4872-8508-059df805e132'
    url = 'https://pro-api.coinmarketcap.com/v1/cryptocurrency/listings/latest'
    parameters = {
        'start': '1',
        'limit': '15',
        'convert': 'USD'
    }
    headers = {
        'Accepts': 'application/json',
        'X-CMC_PRO_API_KEY': '6fae45d7-aea7-4872-8508-059df805e132',
    }

    try:
        # Fetch data from the API
        response = requests.get(url, params=parameters, headers=headers)
        response.raise_for_status()
        data = response.json()['data']
        df = pd.json_normalize(data)
        df['timestamp'] = pd.to_datetime('now', utc=True)
    except (requests.exceptions.RequestException, KeyError) as e:
        print(f"Error fetching data: {e}")
```

```python
# Set up email parameters
sender_email = "willstevens2000@hotmail.co.uk"
receiver_email = "willstevens2000@hotmail.co.uk"
password = "############" #Remeber to remove

# Set the threshold for a 7% drop
threshold_percent = -7

while True:
    # Fetch data from the API
    api_runner()

    # Check if the DataFrame is not empty
    if not df.empty:
        # Calculate the threshold price for each coin
        df['threshold_price'] = df['quote.USD.price'] * (1 + (threshold_percent / 100))

        # Filter coins that are currently below the threshold
        alert_coins = df[df['quote.USD.percent_change_24h'] < threshold_percent]

        # Check if there are alert coins
        if not alert_coins.empty:
            # Create message object instance
            msg = MIMEMultipart()
            msg['From'] = sender_email
            msg['To'] = receiver_email
            msg['Subject'] = "Price Drop Alert"

            # Compose the email content
            email_content = """
                Price Drop Alert!\n
                The following coins have experienced a drop of more than 7% in the last 24 hours:\n
            """ + alert_coins[['name', 'quote.USD.price', 'quote.USD.percent_change_24h', 'threshold_price']].to_string(index=False)

            msg.attach(MIMEText(email_content, 'plain'))

            # Connect to the SMTP server
            server = smtplib.SMTP('smtp.office365.com', 587)
            server.starttls()

            try:
                # Login to the email account
                server.login(sender_email, password)

                # Send email
                server.sendmail(sender_email, receiver_email, msg.as_string())

                print("Email sent successfully!")

            except Exception as e:
                print(f"An error occurred: {e}")

            finally:
                # Close the SMTP server connection
                server.quit()

    # Wait for one hour before running the loop again
    sleep(3600)
```
